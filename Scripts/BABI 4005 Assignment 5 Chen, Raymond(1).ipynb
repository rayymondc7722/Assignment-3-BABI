```python
import pandas as pd

# Load datasets
regular_season = pd.read_excel("../Data/NBA Regular Season 2023-2024 Player Statistics V2.xlsx")
playoffs = pd.read_csv("../Data/NBA Stats 202324 All Stats NBA Player Props Tool (4).csv")

# Standardize column names to match for merging
regular_season.rename(columns={"Player": "NAME"}, inplace=True)

# Merge datasets on 'NAME' and 'TEAM'
merged_data = pd.merge(regular_season, playoffs, on=["NAME"], how="inner", suffixes=("_Regular", "_Playoffs"))

# Fill all NaN values with 0
merged_data.fillna(0, inplace=True)

# Save the merged dataset
merged_data.to_csv("merged_nba_stats.csv", index=False)

# Display first few rows of merged data
print(merged_data.head())
```
